```python
artistas_favoritos = ["Michael Jackson", "Martin Garrix", "Rafael Lechowski"] #creamos la lista para luego hacerle un append de la función de get top tracks para la lista nuestra que hemos itinerado

#para obtener el top de canciones necesitaremos el id del artista que obtendremos realizando una búsqueda con código, y una función que nos de el top de canciones
def get_top_tracks(artist):
    searchx = sp.search(q=artist, type="artist", limit=1) #la q es la query o búsqueda y con el tipo especificamos que solo queremos resultados de artistas
    artist_id = searchx ["artists"]["items"][0]["id"] #obtenemos el id especificando artists y items
    top_tracks = sp.artist_top_tracks(artist_id) #aquí obtenemos las canciones mediante la función artist_top_tracks(id)
    return [track["name"] for track in top_tracks["tracks"][:5]] #por último pedimos a la función que nos devuelva una canción(track) de los ["tracks"] de la variable top_tracks
    
top_tracks_list = []
for artistas in artistas_favoritos:
    top_tracks = get_top_tracks(artistas)
    top_tracks_list.append(top_tracks)
    print()
```
